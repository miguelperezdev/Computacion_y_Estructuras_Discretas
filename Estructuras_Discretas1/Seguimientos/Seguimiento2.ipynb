\begin{aligned}
 \quad A \cap (B \cup C) = (A \cap B) \cup (A \cap C)
 \end{aligned}

---

\begin{aligned}
A \cap (B \cup C) &= \{ x \mid x \in A \wedge x \in (B \cup C) \} \\
&= \{ x \mid x \in A \wedge (x \in B \vee x \in C) \} \\
&= \{ x \mid (x \in A \wedge x \in B) \vee (x \in A \wedge x \in C) \} \\
&= \{ x \mid x \in (A \cap B) \vee x \in (A \cap C) \} \\
&= (A \cap B) \cup (A \cap C)
\end{aligned}

\begin{aligned}
(A - C) \cap (C - B) &= \{ x \mid x \in A \wedge x \notin C \} \cap \{ x \mid x \in C \wedge x \notin B \} \\
&= \{ x \mid x \in A \wedge x \in C \wedge x \notin C \wedge x \notin B \} \\
&= \{ x \mid x \in A \wedge x \notin B \wedge (x \in C \wedge x \notin C) \} \\
&= \{ x \mid x \in A \wedge x \notin B \wedge \emptyset \} \\
&= \emptyset
\end{aligned}

(B - A) \cup (C - A) = (B \cup C) - A
\end{aligned}


---

\begin{aligned}
(B - A) \cup (C - A) &= \{ x \mid x \in B \wedge x \notin A \} \cup \{ x \mid x \in C \wedge x \notin A \} \\
&= \{ x \mid (x \in B \wedge x \notin A) \vee (x \in C \wedge x \notin A) \} \\
&= \{ x \mid x \in (B \cup C) \wedge x \notin A \} \\
&= (B \cup C) - A
\end{aligned}
